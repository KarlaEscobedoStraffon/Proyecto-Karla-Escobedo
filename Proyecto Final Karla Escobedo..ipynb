Posterior a analizar un poco la acción de Apple, elegiremos otras del mercado accionario para hacer un análisis mas completo. Se usarán Apple, LG y Microsoft. Se usará el mismo periodo de tiempo.

```python
start_date= '2019-11-01'
end_date= '2021-11-01'
tickers_list = ['AAPL', 'MSFT', 'LPL']
stocks = pd.DataFrame(columns=tickers_list)
for ticker in tickers_list:
    stocks[ticker] = yf.download(ticker, start_date, end_date) ['Adj Close']
stocks.head()
```

```python
stocks.tail()
```

```python
norm=stocks.div(stocks.iloc[0]).mul(150)
norm.plot(subplots=False)
plt.axhline(y=0, color='black', linestyle='--', linewidth=1)
plt.show()
```

Continuemos con nuestra primera acción analizada que es Apple, lo que haremos en este momento es usar la media movil y la desviación estándar para finalizar con este documento.

MEDIA MOVIL:

```python
aapl = stocks["AAPL"]
aapl = aapl.to_frame()
aapl = aapl.fillna(method="bfill")
aapl["MA30"] = aapl["AAPL"].rolling(window=30).mean()
aapl["MA90"] = aapl["AAPL"].rolling(window=90).mean()
aapl["MA180"] = aapl["AAPL"].rolling(window=180).mean()
aapl["MA360"] = aapl["AAPL"].rolling(window=360).mean()
```

```python
plt.style.use("default")
ax = aapl["AAPL"].plot()
ax.plot(aapl["MA30"], linestyle="--", label="MA30")
ax.plot(aapl["MA90"], linestyle="--", label="MA90")
ax.plot(aapl["MA180"], linestyle="--", label="MA180")
ax.plot(aapl["MA360"], linestyle="--", label="MA360")
ax.legend(loc = "upper left")
ax.set_title("Apple Stock Price")
ax.set_ylabel("Price")
ax.set_xlabel("Date")
plt.show()
```

DESVIACIÓN ESTANDAR:

```python
ax2 = aapl["AAPL"].rolling("30D").agg(["mean", "std"]).plot(subplots=True)
ax2[0].set_title("Apple Stock Price")
ax2[0].set_ylabel("Mean Price")
ax2[1].set_ylabel("Standard Deviation")
plt.show()
```
